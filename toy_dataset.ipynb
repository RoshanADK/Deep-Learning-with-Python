```python
# Generate toy dataset
import numpy as np
import pylab as pl
```

```python
# input
x = np.linspace(-1,1,100)
```

```python
# Based on 2x^square + x + 2
signal =  2 * x * x + x + 2
```

```python
# Lets define some noise(100) from normal distribution with mean=0 and SD=0.1
noise = np.random.normal(0,0.1,100)
```

```python
# Adding up all the noise and signal values
y = signal + noise
```

```python
pl.plot(signal, 'b')
pl.plot(y,'r')
pl.plot(noise, 'g')
pl.xlabel("x")
pl.ylabel("y")
pl.legend(["Without Noise", "With Noise", "Noise"], loc = 2)
x_train = x[0:80]
y_train = y[0:80]

pl.show()
pl.figure()

```

# Now select which model you want to test. Choose one. 

# Models Creation

# Model No. 1 (Degree 1)

```python
# Model with degree 2 defined 
```

```python
# Degree 2 model
degree = 2
x_train = np.column_stack([np.power(x_train,i) for i in range(0,degree)])

# Define Least Square Model as Beta = ( X^Transpose X ) X^Transpose y.
model = np.dot(np.dot(np.linalg.inv(np.dot(x_train.transpose(),x_train)), x_train.transpose()), y_train)

print(np.shape(x_train))
print(np.shape(y_train))
pl.plot(x,y,'g')
pl.xlabel("X")
pl.ylabel("y")
```

```python
predicted = np.dot(model, [np.power(x,i) for i in range(0,degree)])
pl.plot(x,predicted, 'r')
pl.legend(["Actual", "Predicted"], loc = 2)
```

## Error

```python
train_rmse1 = np.sqrt(np.sum(np.dot(y[0:80] - predicted[0:80], y_train - predicted[0:80])))
```

```python
test_rmse1 = np.sqrt(np.sum(np.dot(y[80:] - predicted[80:], y[80:]-predicted[80:])))
```

```python
print("Train RMSE (Degree 1)", train_rmse1)
```

```python
print("Test RMSE (Degree 1)", test_rmse1)
```

# Model No. 2 (Degree 2)

```python
# Degree 2 model
degree = 3
x_train = np.column_stack([np.power(x_train,i) for i in range(0,degree)])

# Define Least Square Model as Beta = ( X^Transpose X ) X^Transpose y.
model = np.dot(np.dot(np.linalg.inv(np.dot(x_train.transpose(),x_train)), x_train.transpose()), y_train)

print(np.shape(x_train))
print(np.shape(y_train))

pl.plot(x,y,'g')
pl.xlabel("X")
pl.ylabel("y")

predicted = np.dot(model, [np.power(x,i) for i in range(0,degree)])
pl.plot(x,predicted, 'r')
pl.legend(["Actual", "Predicted"], loc = 2)
```

## Error

```python
train_rmse1 = np.sqrt(np.sum(np.dot(y[0:80] - predicted[0:80], y_train - predicted[0:80])))
test_rmse1 = np.sqrt(np.sum(np.dot(y[80:] - predicted[80:], y[80:]-predicted[80:])))
print("Train RMSE (Degree 2)", train_rmse1)
print("Test RMSE (Degree 2)", test_rmse1)
```
